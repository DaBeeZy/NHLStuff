```python
import pandas as pd
import numpy as np
import requests

filepath = r'C:\Users\Moe\Documents\Python\NHLanalytics\Rosters\2020\\'
teamCodes = ['ducks','coyotes','bruins','sabres','flames','hurricanes','blackhawks','avalanche','bluejackets','stars','redwings','oilers','panthers','kings','wild','canadiens','predators','devils','islanders','rangers','senators','flyers','penguins','sharks','blues','lightning','mapleleafs','canucks','goldenknights','capitals','jets']

for i in range(len(teamCodes)):
    
    fullroster = pd.DataFrame(columns = ['C','LW','RW','LD','RD','G'])
    team = teamCodes[i]
    url = 'https://www.capfriendly.com/depth-charts/'+team
    html = requests.get(url).content
    dfList = pd.read_html(html)

    fTable = dfList[0].loc[:,['POS','FORWARDS']].dropna().drop(0).sort_values(by='POS')
    fTable.set_index('POS',inplace=True)

    dTable = dfList[1].loc[:,['POS','DEFENSE']].dropna().drop(0).sort_values(by='POS')
    dTable.set_index('POS',inplace=True)

    gTable = dfList[2].loc[:,['POS','GOALIES']].dropna().drop(0)
    gTable.set_index('POS',inplace=True)

    fullroster['C'] = fTable.at['C','FORWARDS']
    fullroster['LW'] = fTable.at['LW','FORWARDS']
    fullroster['RW'] = fTable.at['RW','FORWARDS']
    fullroster['LD'] = pd.Series(dTable.at['LD','DEFENSE'])
    fullroster['RD'] = pd.Series(dTable.at['RD','DEFENSE'])
    fullroster['G'] = pd.Series(gTable.at['G','GOALIES'])
    fullroster = fullroster.fillna('')
    fullroster.index.names = ['Line']
    fullroster.index = fullroster.index + 1
    fullroster.to_csv(filepath+'2020 '+team+' Roster.csv')
    display(fullroster)
    print(str(i+1)+'/31')

print('done')
```
